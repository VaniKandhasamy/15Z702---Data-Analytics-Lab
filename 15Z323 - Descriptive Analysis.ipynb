# Ex1 - Descriptive Analysis

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as mpl
data=pd.read_csv("C:/Users/Dheepak ram/Desktop/EmployeeAttrition.csv")
data.head()
```

It reads the csv file and converts it into a dataframe and displays the first five records in the dataset.

```python
data.describe()
```

It summarises or returns the statistics about the numerical columns in a dataset.

```python
data.sum()
```

It returns the sum of all the columns in the dataset.

```python
data.groupby('Attrition').mean()
```

It groups by the attrition value and returns all the mean values for all the other columns.Based on this we can infer that person with less environment satisfaction will have an attrition value of 'yes'.

```python
data.groupby('Attrition').size()
```

It gives the count of people with the attrition value of 'yes' and 'no'.Based on this we can infer that there are less number of peoples with the attrition value of 'yes' when compared to 'no'. 

```python
data.groupby(['Age','Attrition']).size()
```

It gives the count of people with attrition value of 'yes' and 'no' based on their age.Based on this we can infer that at the age of 25 to 40 the attrition value of 'yes' is more when compared to all the other ages.

```python
data.isnull().sum(axis=0)
```

It returns the count of missing values that is present in each column. From this we can say that there are no missing values in the given csv file.

```python
data.boxplot('Age')
```

It returns the boxplot for the given attribute.From this we can know that the minimum value of age is 18,maximum value is 60 and 25% is 30,median is 35% and 75% is 44.

```python
pd.crosstab(data.MaritalStatus,data.Attrition).plot(kind='bar')
mpl.title('Graph')
mpl.xlabel('MaritalStatus')
mpl.ylabel('Attrition')
```

It displays a bar graph where the x axis is MaritalStatus and y axis is Attrition.Based on this we can infer that the married have the highest attrition value of 'no'when compared to divorced and single.
