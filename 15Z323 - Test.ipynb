```python
import pandas
df = pandas.read_csv('Z:\dataset.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
cst= pandas.read_csv("Z:\dataset.csv", sep=',', usecols=['cst'], squeeze=True)
g=0
obc=0
mobc=0
st=0
sc=0
print("Caste")
for i in cst:
    if i=="G":
        g+=1
    if i=="OBC":
        obc+=1
    if i=="MOBC":
        mobc+=1
    if i=="ST":
        st+=1
    if i=="SC":
        sc+=1
        

x=['G', 'OBC', 'MOBC','ST','SC']
y=[g,obc,mobc,st,sc]
fig,ax=plt.subplots()
ax.bar(x,y, width=0.8, color='r')
plt.show()
print(y)
```

```python
tnp= pandas.read_csv("Z:\dataset.csv", sep=',', usecols=['tnp'], squeeze=True)
vg=0
g=0
p=0
b=0

print("TNP")
for i in tnp:
    if i=="Good":
        g+=1
    if i=="Vg":
        vg+=1
    if i=="Pass":
        p+=1
    if i=="Best":
        b+=1
        

x=['Pass', 'Good', 'Very Good','Best']
y=[p,g,vg,b]
fig,ax=plt.subplots()
ax.bar(x,y, width=0.8, color='g')
plt.show()
print(y)
```
