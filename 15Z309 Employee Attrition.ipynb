```python
import pandas as pd
#function to import the EmployeeAttrition csv file
d = pd.read_csv("EmployeeAttrition.csv")
d.head(10)
```

```python
#This function is used to obtain the minimum value in a column.
d.min()
```

```python
#This function is used to obtain the maximum value in a column.
d.max()
```

```python
#This function is used to display the values of the aggregate functions of all columms
d.describe()
```

```python
#This function is used to generate the scatter plot for the dataset
import matplotlib.pyplot as plb
plb.scatter(d.YearsAtCompany,d.StandardHours, s = d.Education)
#From this plot, we can infer a relation between YearsAtCompany and StandardHours
```

```python
#This function is used to generate the boxplot
d.boxplot(by="WorkLifeBalance", column="Age")
#From this plot, we can infer that the employees with WorkLifeBalance=1 have an average age of 33 and so on.
```

```python
#This function is used to generate the bar chart
d.plot(kind="bar",x="YearsAtCompany",y="WorkLifeBalance")
#From this plot, we can infer that the WorkLifeBalance is more when YearsAtCompany is lesser.
```
