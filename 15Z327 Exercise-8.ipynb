```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
import numpy as np
```

```python
def plot_dendrogram(model, **kwargs):
    c = model.children_
    d = np.arange(c.shape[0])
    n = np.arange(2, c.shape[0]+2)
    matrix = np.column_stack([c,d,n]).astype(float)
    dendrogram(matrix, **kwargs)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
m = AgglomerativeClustering(linkage="ward",n_clusters=3)
sl= m.fit(X)

plt.figure(figsize=(10,10))
plt.title('Single Linkage')
plot_dendrogram(m, labels=sl.labels_)
plt.show()
```

```python
m = AgglomerativeClustering(linkage="complete",n_clusters=3)
cl = m.fit(X)

plt.figure(figsize=(10,10))
plt.title('Complete Linkage')
plot_dendrogram(m, labels=cl.labels_)
plt.show()
```

```python
m = AgglomerativeClustering(linkage="average",n_clusters=3)
al = m.fit(X)

plt.figure(figsize=(10,10))
plt.title('Hierarchical Clustering: Average Linkage')
plot_dendrogram(m, labels=al.labels_)
plt.show()
```
