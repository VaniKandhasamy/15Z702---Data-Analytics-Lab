```python
import urllib
import urllib.request
import pandas as pan

read = pan.read_csv("Y:\DA LAB\MovieGenre.csv", encoding = 'ISO-8859-1')
poster_url = read.Poster
count = 0

for i in poster_url :
    try:
        urllib.request.urlretrieve(i, "Z:\Academics\Semester VII\Data Analytics Laboratory\MovieGenre" + str(count) + ".jpg")
        count += 1
    except:
        print(":( Missing or corrupted link.")
```
