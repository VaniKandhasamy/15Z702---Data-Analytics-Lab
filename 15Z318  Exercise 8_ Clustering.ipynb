```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
iris = load_iris()
```

```python
x = iris.data
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(x)
```

```python
def plot_dendrogram(model, **kwargs):
    child = model.children_
    dist = np.arange(child.shape[0])
    observations = np.arange(2, child.shape[0]+2)
    linkagemat = np.column_stack([child, dist, observations]).astype(float)
    dendrogram(linkagemat, **kwargs)
```

```python
model = AgglomerativeClustering(linkage="ward", n_clusters = 5)
singleLinkage = model.fit(x)

plt.figure(figsize=(15,15))
plt.title('Single Linkage')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plot_dendrogram(model, labels = singleLinkage.labels_)
plt.show()
```

```python
model = AgglomerativeClustering(linkage="complete", n_clusters = 5)
completeLinkage = model.fit(x)

plt.figure(figsize=(15,15))
plt.title('Complete Linkage')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plot_dendrogram(model, labels = completeLinkage.labels_)
plt.show()
```

```python
model = AgglomerativeClustering(linkage="average", n_clusters = 5)
averageLinkage = model.fit(x)

plt.figure(figsize=(15,15))
plt.title('Average Linkage')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plot_dendrogram(model, labels = averageLinkage.labels_)
plt.show()
```
