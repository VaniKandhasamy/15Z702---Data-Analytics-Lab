```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

d = load_iris()
X = d.data

kmean = KMeans(n_clusters=5, random_state=0).fit(X)
kmean.labels_
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
import numpy as np
def plot_dendrogram(model, **kwargs):
    child = model.children_
    d = np.arange(child.shape[0])
    n = np.arange(2, child.shape[0]+2)
    matrix = np.column_stack([child,d,n]).astype(float)
    dendrogram(matrix, **kwargs)
```

```python
from sklearn.cluster import AgglomerativeClustering
fig= AgglomerativeClustering(linkage="ward",n_clusters=3)
single= fig.fit(X)

plt.figure(figsize=(15,10))
plt.title('Single_Linkage')
plot_dendrogram(fig, labels=single.labels_)
plt.show()
```

```python
fig = AgglomerativeClustering(linkage="complete",n_clusters=3)
complete = fig.fit(X)

plt.figure(figsize=(10,10))
plt.title('Complete_Linkage')
plot_dendrogram(fig, labels=complete.labels_)
plt.show()
```

```python
fig = AgglomerativeClustering(linkage="average",n_clusters=3)
average = fig.fit(X)

plt.figure(figsize=(15,10))
plt.title('Average_Linkage')
plot_dendrogram(fig, labels=average.labels_)
plt.show()
```
