```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
import numpy as np
from sklearn.datasets import load_iris
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
print("--AGGLOMERATIVE CLUSTERING--")
dataset = load_iris()
x = dataset.data
model = AgglomerativeClustering(linkage="ward",n_clusters=3)
print("SINGLE LINKAGE")
singleLinkage = model.fit(x)
dendro(model, labels=singleLinkage.labels_)
plt.show()
print("COMPLETE LINKAGE")
model = AgglomerativeClustering(linkage="complete",n_clusters=3)
completeLinkage = model.fit(x)
dendro(model, labels=completeLinkage.labels_)
plt.show()
print("AVERAGE LINKAGE")
model = AgglomerativeClustering(linkage="average",n_clusters=3)
averageLinkage = model.fit(x)
dendro(model, labels=averageLinkage.labels_)
plt.show()
def dendro(model, **kwargs):
    children = model.children_
    distance = np.arange(children.shape[0])
    total = np.arange(2, children.shape[0]+2)
    linkage_matrix = np.column_stack([children, distance, total]).astype(float)
    dendrogram(linkage_matrix, **kwargs)

```
