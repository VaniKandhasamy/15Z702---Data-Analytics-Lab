```python
import numpy as np
import pandas as pd
df=pd.read_csv("movie_metadata.csv")
df.head
```

```python
df=df[['num_critic_for_reviews','director_facebook_likes','actor_3_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','imdb_score']]
df.head()
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    if df[i].isnull().sum()>0:
        df[i]=df[i].fillna(df[i].mean())
    
```

```python
from sklearn.neural_network import MLPClassifier
Y=df['imdb_score'].round().values
X=df.drop(['imdb_score'],axis=1).values
classifier=MLPClassifier(hidden_layer_sizes=5,solver='sgd',activation="logistic")

from sklearn.model_selection import KFold
k=KFold(n_splits=10)
for train,test in k.split(X):
    classifier.fit(X[train],Y[train])
    print(classifier.score(X[test],Y[test]))
```
